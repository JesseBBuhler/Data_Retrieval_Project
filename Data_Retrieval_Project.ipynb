```python
#https://rawg.io/apidocs
#I'm using a video game database API

import pandas as pd
import requests
import json

#creating a blank dataframe
df = pd.DataFrame(columns=["id", "name", "slug", "rating", "playtime", "cover_image_link", "description", "esrb_rating" ,"genre", "reviews_count", "num_platforms", "released", "creators_count"])
df.set_index("id")

#putting all of the elements of the request into variables
search = "games"
page = 1
url = f"https://api.rawg.io/api/{search}"
key = "93f5a48810c34c2bb63496f9da50645e"
keyquery = "?key=" + key

#I want to retrieve 1000 records and there are 20 records in each page so I want to make 50 requests
for page in range(1, 51): #base 1 not base 0
  print("page: " + str(page))
  #making the request
  response = requests.request("GET", url + keyquery + "&page=" + str(page))

  #loading the request into a json object
  df_json = json.loads(response.text)

  #making a request for the details of each game in this page of the response
  for result in df_json["results"]:
    id = result["id"]
    print("ID: " + str(id))

    #making request for the given game ID
    game_response = requests.request("GET", url + "/" + str(id) + keyquery)
    game_json = json.loads(game_response.text)

    # adding the fields from the response to the dataframe
    id = game_json["id"]
    name = game_json["name"]
    slug = game_json["slug"]
    rating = game_json["rating"]
    playtime = game_json["playtime"]
    cover_image_link = game_json["background_image"]
    description = game_json["description_raw"]
    if game_json["esrb_rating"] and "slug" in game_json["esrb_rating"]:
      esrb_rating = game_json["esrb_rating"]["slug"]
    else:
      esrb_rating = ""
    if game_json["genres"] and len(game_json["genres"]) > 0 and game_json["genres"][0] and "slug" in game_json["genres"][0]:
      genre = game_json["genres"][0]["slug"]
    else:
      genre = ""
    reviews_count = game_json["reviews_count"]
    num_platforms = len(game_json["platforms"])
    released = game_json["released"]
    creators_count = game_json["creators_count"]

    df.loc[id] = [id, name, slug, rating, playtime, cover_image_link, description, esrb_rating, genre, reviews_count, num_platforms, released, creators_count]

#output the dataframe
df
```

```python
#saving dataframe to a csv
df.to_csv('/content/drive/MyDrive/School/Winter 2023/Machine Learning/Data_Retrieval_Project/project_dataframe.csv')
```

```python
#saving image URLs into my google drive
for index, row in df.iterrows():
  file_name = row["slug"]
  image_res = requests.get(row["cover_image_link"])
  image = open(f'/content/drive/MyDrive/School/Winter 2023/Machine Learning/Data_Retrieval_Project/Images/{file_name}.jpg', "wb")
  image.write(image_res.content)
  image.close()
```
